```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample population data
data = {
    'Country': ['India', 'China', 'USA', 'Indonesia', 'Pakistan', 'Brazil', 'Nigeria', 'Bangladesh', 'Russia', 'Mexico'],
    'Population': [1417173173, 1425887337, 339996563, 277534122, 240485658, 215353593, 223804632, 172452773, 144373535, 128455567]
}

df = pd.DataFrame(data)

# Colors for each bar
colors = ['#FF9999','#66B2FF','#99FF99','#FFCC99','#CCCCFF','#FFB266','#B2FF66','#FF66B2','#66FFCC','#FF6666']

# Horizontal bar chart
plt.figure(figsize=(12, 6))
plt.barh(df['Country'], df['Population'], color=colors)
plt.title('Population Distribution by Country (Horizontal Bar Chart)', fontsize=14)
plt.xlabel('Population')
plt.ylabel('Country')
plt.tight_layout()

# Save the chart
plt.savefig('population_distribution_horizontal_chart.png')
plt.show()
```
